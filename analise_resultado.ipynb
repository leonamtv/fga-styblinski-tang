# Analisando os resultados com crossover aritm√©tico

```python
import pandas as pnd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
import os

sns.set()
```

```python
resultado_ca = './resultado/ca.csv'
resultado_blx = './resultado/blx.csv'

charts_folder = 'analise'

if not os.path.isdir(charts_folder) :
    os.makedirs(charts_folder)
```

```python
df_ca = pnd.read_csv(resultado_ca)
df_blx = pnd.read_csv(resultado_blx)
df_blx['Num'] = df_blx['Num'] - 85
df_ca['tempo'] = df_ca['tempo'].str.replace(',', '.')
df_ca['avaliacao'] = df_ca['avaliacao'].str.replace(',', '.')
df_ca['min'] = df_ca['min'].str.replace(',', '.')
df_ca['erro'] = df_ca['erro'].str.replace(',', '.')
df_blx['tempo'] = df_blx['tempo'].str.replace(',', '.')
df_blx['avaliacao'] = df_blx['avaliacao'].str.replace(',', '.')
df_blx['min'] = df_blx['min'].str.replace(',', '.')
df_blx['erro'] = df_blx['erro'].str.replace(',', '.')
df_ca = df_ca.astype({ 'tempo' : 'float32', 'avaliacao' : 'float32', 'min' : 'float32', 'erro' : 'float32' })
df_blx = df_blx.astype({ 'tempo' : 'float32', 'avaliacao' : 'float32', 'min' : 'float32', 'erro' : 'float32' })
```

```python
fig, axs = plt.subplots(2, 3, figsize=(20, 10),constrained_layout=True)

fig.suptitle('Taxa de muta√ß√£o x tempo', size=18)

axs[0, 0].set_title('[ nPop=20, nGer=10000, nElite=2, dim=2 ]')
axs[0, 0].set_xlabel('taxa de muta√ß√£o')
axs[0, 0].set_ylabel('tempo(s)')
axs[0, 0].set_xticks(df_ca[:5]['tM'])
ax2 = axs[0, 0].twinx()
ax2.grid(False)
sns.lineplot( data = df_ca[:5], x='tM', y='tempo', linewidth=2.5, label='tempo ca', ax=axs[0, 0])
sns.lineplot( data = df_blx[:5], x='tM', y='tempo', linewidth=2.5, label='tempo blx', ax=axs[0, 0])
sns.lineplot( data = df_blx[:5], x='tM', y='erro', linewidth=2.5,  label='erro blx', ax=ax2, color='red')
sns.lineplot( data = df_ca[:5], x='tM', y='erro', linewidth=2.5, label='erro ca', ax=ax2, color='green')
axs[0, 0].minorticks_off()
h1, l1 = axs[0, 0].get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax2.get_legend().remove()
axs[0, 0].legend(h1+h2, l1+l2, loc=2)

axs[0, 1].set_title('[ nPop=20, nGer=10000, nElite=2, dim=4 ]')
axs[0, 1].set_xlabel('taxa de muta√ß√£o')
axs[0, 1].set_ylabel('tempo(s)')
axs[0, 1].set_xticks(df_ca[14:19]['tM'])
ax3 = axs[0, 1].twinx()
ax3.grid(False)
sns.lineplot( data = df_ca[14:19], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo ca', ax=axs[0, 1])
sns.lineplot( data = df_blx[14:19], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo blx', ax=axs[0, 1])
sns.lineplot( data = df_blx[14:19], x='tM', y='erro', linewidth=2.5, legend='full', label='erro blx', ax=ax3, color='red')
sns.lineplot( data = df_ca[14:19], x='tM', y='erro', linewidth=2.5, legend='full', label='erro ca', ax=ax3, color='green')
axs[0, 1].minorticks_off()
h1, l1 = axs[0, 1].get_legend_handles_labels()
h2, l2 = ax3.get_legend_handles_labels()
ax3.get_legend().remove()
axs[0, 1].legend(h1+h2, l1+l2, loc=2)

axs[0, 2].set_title('[ nPop=20, nGer=10000, nElite=2, dim=6 ]')
axs[0, 2].set_xlabel('taxa de muta√ß√£o')
axs[0, 2].set_ylabel('tempo(s)')
axs[0, 2].set_xticks(df_ca[28:33]['tM'])
ax4 = axs[0, 2].twinx()
ax4.grid(False)
sns.lineplot( data = df_ca[28:33], x='tM', y='tempo', linewidth=2.5, label='tempo ca', ax=axs[0, 2])
sns.lineplot( data = df_blx[28:33], x='tM', y='tempo', linewidth=2.5, label='tempo blx', ax=axs[0, 2])
sns.lineplot( data = df_blx[14:19], x='tM', y='erro', linewidth=2.5, label='erro blx', ax=ax4, color='red')
sns.lineplot( data = df_ca[14:19], x='tM', y='erro', linewidth=2.5, label='erro ca', ax=ax4, color='green')
axs[0, 2].minorticks_off()
h1, l1 = axs[0, 2].get_legend_handles_labels()
h2, l2 = ax4.get_legend_handles_labels()
ax4.get_legend().remove()
axs[0, 2].legend(h1+h2, l1+l2, loc=2)

axs[1, 0].set_title('[ nPop=20, nGer=10000, nElite=2, dim=8 ]')
axs[1, 0].set_xlabel('taxa de muta√ß√£o')
axs[1, 0].set_ylabel('tempo(s)')
axs[1, 0].set_xticks(df_ca[42:47]['tM'])
ax5 = axs[1, 0].twinx()
ax5.grid(False)
sns.lineplot( data = df_ca[42:47], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo ca', ax=axs[1, 0])
sns.lineplot( data = df_blx[42:47], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo blx', ax=axs[1, 0])
sns.lineplot( data = df_blx[42:47], x='tM', y='erro', linewidth=2.5, legend='full', label='erro blx', ax=ax5, color='red')
sns.lineplot( data = df_ca[42:47], x='tM', y='erro', linewidth=2.5, legend='full', label='erro ca', ax=ax5, color='green')
axs[1, 0].minorticks_off()
h1, l1 = axs[1, 0].get_legend_handles_labels()
h2, l2 = ax5.get_legend_handles_labels()
ax5.get_legend().remove()
axs[1, 0].legend(h1+h2, l1+l2, loc=2)

axs[1, 1].set_title('[ nPop=20, nGer=10000, nElite=2, dim=10 ]')
axs[1, 1].set_xlabel('taxa de muta√ß√£o')
axs[1, 1].set_ylabel('tempo(s)')
axs[1, 1].set_xticks(df_ca[56:61]['tM'])
ax6 = axs[1, 1].twinx()
ax6.grid(False)
sns.lineplot( data = df_ca[56:61], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo ca', ax=axs[1, 1])
sns.lineplot( data = df_blx[56:61], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo blx', ax=axs[1, 1])
sns.lineplot( data = df_blx[56:61], x='tM', y='erro', linewidth=2.5, legend='full', label='erro blx', ax=ax6, color='red')
sns.lineplot( data = df_ca[56:61], x='tM', y='erro', linewidth=2.5, legend='full', label='erro ca', ax=ax6, color='green')
axs[1, 1].minorticks_off()
h1, l1 = axs[1, 1].get_legend_handles_labels()
h2, l2 = ax6.get_legend_handles_labels()
ax6.get_legend().remove()
axs[1, 1].legend(h1+h2, l1+l2, loc=2)

axs[1, 2].sharey = True
axs[1, 2].set_title('[ nPop=20, nGer=10000, nElite=2, dim=12 ]')
axs[1, 2].set_xlabel('taxa de muta√ß√£o')
axs[1, 2].set_ylabel('tempo(s)')
axs[1, 2].set_xticks(df_ca[70:75]['tM'])
ax7 = axs[1, 2].twinx()
ax7.grid(False)
sns.lineplot( data = df_ca[70:75], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo ca', ax=axs[1, 2])
sns.lineplot( data = df_blx[70:75], x='tM', y='tempo', linewidth=2.5, legend='full', label='tempo blx', ax=axs[1, 2])
sns.lineplot( data = df_blx[70:75], x='tM', y='erro', linewidth=2.5, legend='full', label='erro blx', ax=ax7, color='red')
sns.lineplot( data = df_ca[70:75], x='tM', y='erro', linewidth=2.5, legend='full', label='erro ca', ax=ax7, color='green')
axs[1, 2].minorticks_off()
h1, l1 = axs[1, 2].get_legend_handles_labels()
h2, l2 = ax7.get_legend_handles_labels()
ax7.get_legend().remove()
axs[1, 2].legend(h1+h2, l1+l2, loc=2)

plt.savefig(os.path.join(charts_folder, 'tm_vs_tmp.pdf'))
```
