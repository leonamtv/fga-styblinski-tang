```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
tx, ty = np.linspace(-5, 5, 400), np.linspace(-5, 5, 400)
xx, yy = np.meshgrid(tx, ty)

tz = 0.5 * (( xx**4 - 16*xx**2 + 5*xx) + ( yy**4 - 16*yy**2 + 5*yy))
```

```python
plt.figure(figsize=(20, 18))
ax = plt.axes(projection='3d')
ax.contour3D(tx, ty, tz, 200)
plt.savefig('st.svg')
```
